```python
#Importando datos
url = 'https://raw.github.com/finiterank/homicidios/master/data/homicidios.1990.a.2013.csv'
webpage = urllib.urlopen(url)
datareader = csv.reader(webpage)
data = []
for row in datareader:
    data.append(row)
print data
```

```python
# adecuación matriz datos
datah = []
columna = []
for i in range(len(data[0])):
    for k in range(24):
        if (k + 90)%100 >9:
            u = str((k+90)%100)
        else:
            u = str(0) + str((k + 90)%100)
        if data[0][i] == "tas" + u:
            columna = []
            for j in range(len(data)-1):
                columna.append(data[j+1][i])
            datah.append(columna)
print datah         
        
```

```python
# método para determinar covarianza
def cov(a,b):
    covar = 0.0
    amean = 0.0
    bmean = 0.0
    ca = 0.0
    cb = 0.0
    for i in range(len(a)):
        if (a[i] != 'NA' and a[i] != "#DIV/0!"):
            amean = amean + float(a[i])
        else:
            ca = ca +1
        if (b[i] != "NA" and b[i] != "#DIV/0!"):
            bmean = bmean + float(b[i])
        else:
            cb = cb + 1
    amean = amean/(float(len(a))-ca)
    bmean = bmean/(float(len(b))-cb)
    Ma = float(len(a)) - ca - 1
    Mb = float(len(a)) - cb - 1
    for i in range(len(a)):
        if (a[i] != 'NA' and a[i] != "#DIV/0!"):
            if (b[i] != 'NA' and b[i] != "#DIV/0!"):
                covar = covar + (float(a[i])-amean)*(float(b[i])-bmean)/min(Ma,Mb)
    return covar

```

```python
#Determinación de autovalores y autovectores
l = len(datah)
Matriz = zeros([l,l])
for i in range(l):
    for j in range(l):
        Matriz[i][j]=cov(datah[i],datah[j])
eigen = eig(Matriz)
print Matriz
```
